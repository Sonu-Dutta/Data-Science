```python
# lets create a new column with name Gender
df['Gender'] = df.Sex.map({'male':0,'female':1})
```

```python
# handling of missing values
df.isna().sum()
```

```python
# in age column we have 86 missing entries, lets try to fill it using average age
round(df.Age.mean())
```

```python
df.Age.fillna(round(df.Age.mean()))
```

```python
# now replace above in the same column
df['Age'] = df.Age.fillna(round(df.Age.mean()))
```

```python
#lets check dataframe after filling of missing values, 
# so age is not having any missing value now
df.isna().sum()
```

```python
# Delete unnecessary columns from dataset (Fare,Cabin,Embarked,Sex)
final = df.drop(['Fare','Cabin','Embarked','Sex'],axis=1)
```

```python
# Feature selection:
# Use this sequence: Name,age,gender,SibSp, Parch,Pclass,Ticket,Survived
final = df[['Name','Age','Gender','SibSp', 'Parch','Pclass','Ticket','Survived']]
final[:2]
```
